```python
import pandas as pd 
import matplotlib.pyplot as ptl
import numpy as np
import collections
import datetime
import datetime as dt 
```

```python
dados_covid = pd.read_csv("dados1.csv")
dados_covid.head()
```

```python
X = dados_covid.iloc[:,0].values
X = np.sort(X)
```

```python
Y = collections.Counter(X)
def remove_name(lista):
    l = []
    for i in lista:
            l.append(lista[i])
    return l

Y = remove_name(Y)
Y.reverse()
Y
```

```python
def remove_repetidos(lista):
    l = []
    for i in lista:
        if i not in l:
            l.append(i)
    l.sort()
    return l

Z = remove_repetidos(X)
M = pd.to_datetime(Z).strftime('%Y-%m') 
```

```python
ptl.title("Pacientes de Covid-19")
ptl.xlabel('Fevereiro de 2020 à Abril de 2021')
ptl.ylabel("Numero de Casos")
ptl.plot(Z,Y, label="Casos")
ptl.xticks([], [])
ptl.legend()
ptl.show()
```
